```python
import kagglehub
import pandas as pd
```

```python
path = kagglehub.dataset_download("abdallahalidev/plantvillage-dataset", output_dir="Plant Disease Detection/")
```

```python
import os
data_dir_path = os.path.join(path, "plantvillage dataset", "color")
# Walk through the directory and count images per class (folder)
class_counts = {}

image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff', '.webp'}

for root, dirs, files in os.walk(data_dir_path):
    # Only count at the immediate subdirectory level (classes)
    if os.path.dirname(root) == data_dir_path:
        class_name = os.path.basename(root)
        image_files = [f for f in files if os.path.splitext(f)[1].lower() in image_extensions]
        if image_files:
            class_counts[class_name] = len(image_files)

# Print results
print(f"{'Class':<40} {'Count':>6}")
print("-" * 48)
for cls, count in sorted(class_counts.items()):
    print(f"{cls:<40} {count:>6}")
print("-" * 48)
print(f"{'Total':<40} {sum(class_counts.values()):>6}")
```
